```python
import pandas as pd
df = pd.read_csv("data/citibike-tripdata.csv", sep = ',')
display(df)
```

```python
display(df['gender'].value_counts().nlargest())
```

```python
display(df['start station name'].value_counts().nsmallest())
```

```python
df = df.drop(['start station id', 'end station id'], axis = 1)
display(df.info())
```

```python
df['age'] = df['birth year'].apply(lambda x: 2018-x)
display(df['age'])
```

```python
display(df[df['age']>60])
```

```python
df['starttime'] = pd.to_datetime(df['starttime'], dayfirst=True)
df['stoptime'] = pd.to_datetime(df['stoptime'], dayfirst=True)
df['trip duration'] = df['stoptime']-df['starttime']
display(df['trip duration'].iloc[3])
```

```python
df['weekend'] = df['starttime'].dt.day_of_week.apply(lambda x: 1 if x ==5 or x ==6 else 0)
display(df[df['weekend']==1])
```

```python
def get_time(time):
    if time in range(0, 7): return 'night'
    if time in range(7, 13): return 'morning'
    if time in range(13, 19): return 'day'
    if time in range(19, 24): return 'evening'
df['time_of_day'] = df['starttime'].dt.hour.apply(get_time)
display(df[df['time_of_day']=='night'])
```
